```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(np.random.randn(50, 4), index=pd.date_range('2019-1-1', periods=50), columns=list('ABCD'))
df
```

```python
df.plot(kind='line', style='--o', figsize=(18, 6), grid=True, title='line', colormap='copper')
```

```python
plt.figure(figsize=(18, 6))
plt.scatter(df['A'], df['B'])
```

```python
plt.figure(figsize=(18, 6))
plt.scatter(df['A'], df['B'], s=df['C'].abs() * 100, alpha=0.8, c=df['D'], cmap='Reds')
```

```python
df.boxplot(figsize=(18, 9))
```

```python
sns.set(style='ticks', palette='pastel')
```

```python
plt.figure(figsize=(18, 9))
sns.kdeplot(df['A'], df['B'], shade=True, cmap='Reds', shade_lowest=False)
```
