# Matplotlib数据可视化基本使用方法

```python
# 导入matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 1])
plt.show()
```

三行代码就可通过matplotlib构建一个简单的图形。

在python matplotlib库创建的图形中添加标题，标签。

```python
x = [5, 2, 7]
y = [2, 16, 4]
plt.plot(x, y)
plt.title('Title')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'g',label='line one', linewidth=5)
plt.plot(x2,y2,'r',label='line two',linewidth=5)
plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
#设置图例位置
plt.legend()
plt.grid(True, color='k')
plt.show()
```
